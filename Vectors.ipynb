```python
import numpy as np
import matplotlib.pyplot as plt

vi = float(input("Magnitude of the first vector: "))
vj = float(input("Magnitude of the second vector: "))

xi = vi-3
xf = xi+vi+3

yi = vj-3
yf = yi+vj+3

plt.plot([xi, xi+0.0001],[yf-0.0001, yf], c='white')
plt.plot([xf-0.0001, xf],[yi, yi+0.0001], c='white')
plt.arrow(vi-1, vj-1, vi, vj, head_width = 0.2, width = 0.05, ec ='green')
plt.show();
```

## Vector Addition

```python
import numpy as np
import matplotlib.pyplot as plt

plt.plot([0, 5],[0, 7], c='white')

v1i = 1
v1j = 2
v2i = 2
v2j = 2

plt.arrow(1, 2, v1i, v1j, head_width = 0.2, width = 0.05, ec ='green')
plt.arrow(2, 4, v2i, v2j, head_width = 0.2, width = 0.05, ec ='red')

v3i = v1i + v2i
v3j = v1j + v2j
v3_mag = np.linalg.norm([3, 4])

stuff = r'The magnitude of $\vec{v_3}$' + f' is {v3_mag}'
plt.text(1, 1, stuff, fontsize=14, ha='left')
# print(stuff)
plt.arrow(1, 2, v3i, v3j, head_width = 0.2, width = 0.05, ec ='blue')
plt.show();
```

```python
v1 = np.array([2, 2])
v2 = np.array([1, 2])
v3 = v1 + v2
print(v3)
v3_mag = np.linalg.norm(v3)
print(v3_mag)
```

```python
v4 = 4*v1
v5 = 4*v2
print(v4)
print(v5)
```

```python
v6 = v4 + v5
print(v6)
v6_mag = np.linalg.norm(v6)
print(v6_mag)
```

```python
v1_unit = v1 / np.linalg.norm(v1)
print(f'Unit vector of v1 is {v1_unit}\n')

v2_unit = v2 / np.linalg.norm(v2)
print(f'Unit vector of v2 is {v2_unit}\n')

v3_unit = v3 / v3_mag
print(f'Unit vector of v3 is {v3_unit}\n')

v4_unit = v4 / np.linalg.norm(v4)
print(f'Unit vector of v4 is {v4_unit}\n')

v5_unit = v5 / np.linalg.norm(v5)
print(f'Unit vector of v5 is {v5_unit}\n')

v6_unit = v6 / np.linalg.norm(v6)
print(f'Unit vector of v6 is {v6_unit}\n')
```

```python
print(f'The magnitude of v1_unit is {np.linalg.norm(v1_unit)}\n')
print(f'The magnitude of v2_unit is {np.linalg.norm(v2_unit)}\n')
print(f'The magnitude of v3_unit is {np.linalg.norm(v3_unit)}\n')
print(f'The magnitude of v4_unit is {np.linalg.norm(v4_unit)}\n')
print(f'The magnitude of v5_unit is {np.linalg.norm(v5_unit)}\n')
print(f'The magnitude of v6_unit is {np.linalg.norm(v6_unit)}\n')
```
